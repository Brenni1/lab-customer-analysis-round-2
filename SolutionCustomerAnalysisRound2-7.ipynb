```python
import pandas as pd  
import numpy as np

data = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
display(data.head())
display(data.tail())
data.shape
```

```python
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower())
data.columns = cols
```

```python
pd.set_option('display.max_columns', None)


data[data['state'].isna() & data['response'].isna()]

data[(data['months since last claim'].isna() & data['number of open complaints'].isna())]
```

```python
data['state'] = data['state'].fillna("Unknown")
data['response'] = data['response'].fillna("No")
data['months since last claim'] = data['months since last claim'].fillna(0.0)
data['number of open complaints'] = data['number of open complaints'].fillna(0.0)

data.head(20)
```

```python
data.isna().sum()
```

```python
data['vehicle type'].unique()
```

```python
data['vehicle class'] = data['vehicle class'].fillna("Unknown")
data['vehicle size'] = data['vehicle size'].fillna("Unknown")
data['vehicle type'] = data['vehicle type'].fillna("Unknown")  # change from previous version as suggested, instead of assuming 'A' I changed it to be consistent with the other fills and went for 'Unknown'

# consistency here might come in handy when doing some filtering for only complete datasets.
```
